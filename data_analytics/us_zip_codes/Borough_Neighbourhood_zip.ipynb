```python
import pandas as pd
import json
import numpy as np
```

```python
df = pd.read_csv('nyc-zip-codes.csv')
```

```python
df
```

```python
# Group by Borough and Neighborhood
grouped = df.groupby(['Borough', 'Neighborhood'])['ZipCode'].unique().reset_index()
```

```python
# Convert 
grouped['ZipCode'] = grouped['ZipCode'].apply(lambda x: x.tolist())
```

```python
grouped
```

```python
# dictionary 
result = {}
for borough in grouped['Borough'].unique():
    borough_data = grouped[grouped['Borough'] == borough]
    result[borough] = {row['Neighborhood']: row['ZipCode'] for _, row in borough_data.iterrows()}
```

```python
result
```

```python
# Save 
with open('borough_neighbourhood.json', 'w') as f:
    json.dump(result, f, indent=2)
```
