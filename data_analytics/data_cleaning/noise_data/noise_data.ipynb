## Data for Modelling

```python
import pandas as pd

noise_data = pd.read_csv('cleaned_noise_data.csv')
noise_data
```

```python
noise_data['Incident Zip'] = noise_data['Incident Zip'].astype(str)
```

```python
# Extract year 
noise_data['year'] = pd.to_datetime(noise_data['Created Date']).dt.year
```

```python
# Group by year and zipcode
noise_data_yearly = noise_data.groupby(['year', 'Incident Zip']).size().reset_index(name='num_noise_complaints')
```

```python
noise_data_yearly
```

```python
# Save 
noise_data_yearly.to_csv('noise_model.csv', index=False)
```
