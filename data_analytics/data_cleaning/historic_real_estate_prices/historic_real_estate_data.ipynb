```python
# List of new zip codes and their latest values
new_zipcodes = ['11693', '11005', '10474', '11697', '10475', '11359', '11109', '10038']
latest_values = [499000.00, 384000.67, 434500.00, 1182500, 960000.52, 770000.82, 662500.00, 1100000.00]
```

```python
# Get the date columns (historic prices)
date_columns = [col for col in df_nyc.columns if '-' in col]
```

```python
# Function to calculate the average for each month based on the surrounding zip codes
def calculate_monthly_averages(df, new_zipcodes):
    averages = {}
    for date in date_columns:
        averages[date] = df[date].mean()
    return averages

# Calculate the monthly averages
monthly_averages = calculate_monthly_averages(df_nyc, new_zipcodes)
```

```python
# Create a DataFrame for the new zip codes with the calculated monthly averages
new_data = []
for i, zipcode in enumerate(new_zipcodes):
    row = {'zipcode': zipcode}
    for date in date_columns:
        row[date] = monthly_averages[date]
    row['2024-05-31'] = latest_values[i]  # Override the last month's value
    new_data.append(row)

new_df = pd.DataFrame(new_data)
```

```python
# Concatenate the new DataFrame with the existing filtered DataFrame
df_updated = pd.concat([df_nyc, new_df], ignore_index=True)
```

```python
df_updated
```

```python
missing_values = df_updated.isna().sum()
print(missing_values)
```

```python
# Function to fill NaN values with the average of surrounding zip codes
def fill_missing_values_with_averages(df):
    # Get the list of zip codes in the DataFrame
    zipcodes = df['zipcode'].unique()

    # Iterate over each zip code
    for zipcode in zipcodes:
        # Get the rows for the current zip code
        zip_data = df[df['zipcode'] == zipcode]
        
        # Find rows with missing values
        missing_indices = zip_data[zip_data.isna().any(axis=1)].index

        for idx in missing_indices:
            # Get the date columns (historic prices)
            for date in date_columns:
                if pd.isna(df.at[idx, date]):
                    # Calculate the average from surrounding zip codes
                    surrounding_data = df[(df['zipcode'] != zipcode) & (df[date].notna())][date]
                    df.at[idx, date] = surrounding_data.mean()
    
    return df

# Fill missing values
df_filled = fill_missing_values_with_averages(df_updated)
```

```python
import matplotlib.pyplot as plt
from scipy.interpolate import UnivariateSpline

# Load the updated dataset
df = pd.read_csv('historic_real_estate_price.csv')

# Function to remove outliers based on IQR
def remove_outliers_iqr(df, columns):
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        df = df[(df[col] >= (Q1 - 1.5 * IQR)) & (df[col] <= (Q3 + 1.5 * IQR)) | (df[col].isna())]
    return df

# Function to apply moving average
def moving_average(series, window_size):
    return series.rolling(window=window_size, min_periods=1).mean()

# Get the date columns (historic prices)
date_columns = [col for col in df.columns if '-' in col]

# Remove outliers using IQR
df = remove_outliers_iqr(df, date_columns)

# Function to apply spline smoothing with a fallback to moving average
def apply_smoothing(df, date_columns, spline_s=0.5, ma_window_size=3):
    smoothed_data = []
    for zipcode in df['zipcode'].unique():
        df_zipcode = df[df['zipcode'] == zipcode].copy()
        for date in date_columns:
            y = df_zipcode[date].values
            x = np.arange(len(y))
            
            # Handle NaN or infinite values by interpolation or filling
            if np.any(np.isnan(y)) or np.any(np.isinf(y)):
                y = pd.Series(y).interpolate().fillna(method='bfill').fillna(method='ffill').values
            
            # Apply spline smoothing if enough data points, otherwise use moving average
            if len(y) > 4:  # Minimum number of points required for UnivariateSpline
                try:
                    spline = UnivariateSpline(x, y, s=spline_s)
                    smoothed_values = spline(x)
                except Exception as e:
                    print(f"Spline smoothing failed for zipcode {zipcode} on {date}: {e}")
                    smoothed_values = moving_average(pd.Series(y), ma_window_size).values
            else:
                smoothed_values = moving_average(pd.Series(y), ma_window_size).values
            
            df_zipcode[date] = smoothed_values
        smoothed_data.append(df_zipcode)
    return pd.concat(smoothed_data, ignore_index=True)

# Apply smoothing to each zipcode
df_smoothed = apply_smoothing(df, date_columns, spline_s=0.5, ma_window_size=3)
```
