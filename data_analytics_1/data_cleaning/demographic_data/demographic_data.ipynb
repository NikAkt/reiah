```python
# Function to safely extract values from nested structures
def extract_value(data, key):
    if isinstance(data, list):
        for item in data:
            if item['key'] == 'Data':
                for value in item['values']:
                    if value['x'] == key:
                        return value['y']
    return np.nan
```

```python
# Extract gender information
df_demographic['male_percentage'] = df_demographic['population_by_gender'].apply(lambda x: extract_value(x, 'Male'))
df_demographic['female_percentage'] = df_demographic['population_by_gender'].apply(lambda x: extract_value(x, 'Female'))
```

```python
# Extract race information
race_columns = ['White', 'Black Or African American', 'Asian', 'American Indian Or Alaskan Native', 'Native Hawaiian & Other Pacific Islander', 'Other Race']
for race in race_columns:
    df_demographic[f'race_{race.lower().replace(" ", "_")}'] = df_demographic['population_by_race'].apply(lambda x: extract_value(x, race))
```

```python
# Extract family information
df_demographic['husband_wife_family'] = df_demographic['families_vs_singles'].apply(lambda x: extract_value(x, 'Husband Wife Family Households'))
df_demographic['single_guardian'] = df_demographic['families_vs_singles'].apply(lambda x: extract_value(x, 'Single Guardian'))
df_demographic['singles'] = df_demographic['families_vs_singles'].apply(lambda x: extract_value(x, 'Singles'))
df_demographic['singles_with_roommate'] = df_demographic['families_vs_singles'].apply(lambda x: extract_value(x, 'Singles With Roommate'))
```

```python
# Extract households with kids information
df_demographic['households_without_kids'] = df_demographic['households_with_kids'].apply(lambda x: extract_value(x, 'Households Without Kids'))
df_demographic['households_with_kids'] = df_demographic['households_with_kids'].apply(lambda x: extract_value(x, 'Households With Kids'))
```

```python
def extract_travel_time(data):
    if isinstance(data, list):
        total_time = 0
        total_people = 0
        for item in data:
            if item['key'] == 'Data':
                for value in item['values']:
                    if value['x'] != 'Total':
                        time_range = value['x'].replace('+', '-90').replace('<', '').split('-')
                        try:
                            time_range = list(map(int, time_range))
                        except ValueError:
                            continue  # Skip this iteration if conversion fails
                        avg_time = sum(time_range) / len(time_range)
                        total_time += avg_time * value['y']
                        total_people += value['y']
        return total_time / total_people if total_people > 0 else np.nan
    return np.nan

df_demographic['travel_time_to_work_in_minutes'] = df_demographic['travel_time_to_work_in_minutes'].apply(extract_travel_time)
```

```python
# Drop unnecessary columns
df_demographic = df_demographic.drop(['population_by_gender', 'population_by_race', 'families_vs_singles', 'households_with_kids'], axis=1)
```

```python
def project_data(df, start_year=2018, end_year=2023):
    projected_data = []
    annual_growth_rate = 0.005  
    years_diff = end_year - start_year
    for _, row in df.iterrows():
        zipcode = row['zipcode']
        year_data = {'zipcode': zipcode, 'year': end_year}
        for column in df.columns:
            if column != 'zipcode':
                if pd.api.types.is_numeric_dtype(df[column]):
                    year_data[column] = row[column] * (1 + annual_growth_rate) ** years_diff
                else:
                    year_data[column] = row[column]
        projected_data.append(year_data)

    return pd.DataFrame(projected_data)
```

```python
# Project data 
df_projected = project_data(df_demographic, end_year=2023)
```

```python
# Group household categories
df_projected['FamilyHousehold'] = df_projected['husband_wife_family'] + df_projected['single_guardian']
df_projected['SingleHousehold'] = df_projected['singles'] + df_projected['singles_with_roommate'] + df_projected['households_without_kids']
```

```python
column_mapping = {
    'zipcode': 'ZipCode',
    'year': 'Year',
    'population': 'Population',
    'population_density': 'PopulationDensity',
    'median_household_income': 'MedianHouseholdIncome',
    'travel_time_to_work': 'AverageTravelTimeToWork',
    'male_percentage': 'Male',
    'female_percentage': 'Female',
    'FamilyHouseholdPercentage': 'FamilyHousehold',
    'SingleHouseholdPercentage': 'SingleHousehold',
    'race_white': 'white',
    'race_black_or_african_american': 'black',
    'race_asian': 'asian',
    'race_american_indian_or_alaskan_native': 'american_indian',
    'race_native_hawaiian_&_other_pacific_islander': 'pacific_islander',
    'race_other_race': 'other'
}

df_projected = df_projected.rename(columns=column_mapping)
```

```python
# Calculate Diversity Index
def calculate_diversity_index(row):
    race_columns = ['white', 'black', 'asian', 'american_indian', 'pacific_islander', 'other']
    total_population = sum(row[col] for col in race_columns)
    if total_population == 0:
        return 0
    proportions = [row[col] / total_population for col in race_columns if row[col] > 0]
    return 1 - sum(p * p for p in proportions)

df_projected['DiversityIndex'] = df_projected.apply(calculate_diversity_index, axis=1)
```
